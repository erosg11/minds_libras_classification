```python
import gzip
import regex
import pandas as pd
import numpy as np
```

```python
gp = gzip.open('model_RN_2023-04-24_19-17-20_501117.log.gz', 'rb')
gp
```

```python
data = gp.readlines()
data[:5]
```

```python
data_decoded = [x.decode() for x in data]
data_decoded[:5]
```

```python
data_decoded[:10]
```

```python
header_regex = r'^(?P<date>\d{4}-\d\d-\d\d\s+\d\d:\d\d:\d\d(\.\d*))?\s*\|\s*(?P<level>[^|]*?)\s*\|\s*(?P<módulo>[^:]+):(?P<função>[^:]+):(?P<linha>\d+)\s+-\s+(?P<mensagem>.+)\n$'
header_regex = regex.compile(header_regex)
header_regex
```

```python
headers = [header_regex.match(x) for x in data_decoded]
headers[:10]
```

```python
ixs_none = [i for i, x in enumerate(headers) if x is None]
len(ixs_none)
```

```python
[data_decoded[i] for i in ixs_none]
```

```python
dicts = [x.groupdict() for x in headers if x is not None]
dicts[:10]
```

```python
df = pd.DataFrame(dicts)
df.head()
```

```python
df2 = df.copy()
df2['date'] = pd.to_datetime(df2['date'])
df2['level'] = df2['level'].astype('category')
df2['módulo'] = df2['módulo'].astype('category')
df2['função'] = df2['função'].astype('category')
df2['linha'] = pd.to_numeric(df2['linha'])
df2['mensagem'] = df2['mensagem'].astype('string')
print(df2.info())
df2.head()
```

```python
df3 = df2.copy()
df3['status'] = None
df3.loc[(df3['módulo'] == 'skopt.searchcv') & (df3['linha'] == 408), 'status'] = 'Iniciando fitting'
df3.head(10)
```

```python
df3.loc[(df3['módulo'] == 'KMeansCustomEstimator') & (df3['linha'] == 63), 'status'] = 'Atualizando estimador'
df3[~pd.isnull(df3['status'])].head()
```

```python
df3.loc[(df3['módulo'] == 'KMeansCustomEstimator') & (df3['linha'] == 120), 'status'] = 'Pontuando'
df3[~pd.isnull(df3['status'])].head()
```

```python
mensagem_atualizando = df3.loc[df3['status'] == 'Atualizando estimador', 'mensagem'].values[0]
mensagem_atualizando
```

```python
params = mensagem_atualizando[27:]
params
```

```python
match_kv = regex.finditer(r"'(?P<key>[^']+)': (?P<value><[^>]*>|'[^']*'|False|True|\d+(?:\.\d*)?|[a-zA-Z]*\([^)]*\)|None)+", params)
kv = [x.groupdict() for x in match_kv]
kv
```

```python
mensagem_pontuando = df3.loc[df3['status'] == 'Pontuando', 'mensagem'].values[0]
mensagem_pontuando
```

```python
float(mensagem_pontuando[21:])
```

```python
df4 = df3[~pd.isnull(df3['status'])].reset_index().copy()
df4.head()
```

```python
list_results = {}
current_test = -1
BR = '\n'
curr_dict = None
append_result = None
for i, mensagem, status in df4.loc[:max(df4.index)-1, ['mensagem', 'status']].itertuples(index=True):
    if status == 'Iniciando fitting':
        if curr_dict is not None:
            np_results = np.array(curr_dict['results'])
            curr_dict['mean'] = np_results.mean()
            curr_dict['std'] = np_results.std()
        current_test += 1
        results = []
        curr_dict = {'params': None, 'results': results}
        append_result = results.append
        list_results[current_test] = curr_dict
    elif status == 'Atualizando estimador':
        match_kv = regex.finditer(r"'(?P<key>[^']+)': (?P<value><[^>]*>|'[^']*'|False|True|\d+(?:\.\d*)?|[a-zA-Z]*\([^)]*\)|None)+", mensagem[27:])
        kv = [x.groupdict() for x in match_kv]
        dict_params = {x['key']: x['value'] for x in kv}
        curr_params = curr_dict['params']
        if curr_params is None:
            curr_dict['params'] = dict_params
        elif curr_params != dict_params:
            diffs = set(curr_params.items()) ^ set(dict_params.items())
            raise ValueError(f"""Diferença de parâmetros do index {i}:{BR}{
                BR.join([f'curr_params[{k}] = {curr_params[k]} != dict_parms[{k}] = {dict_params[k]}' for k, v in diffs])}""")
    elif status == 'Pontuando':
        append_result(float(mensagem[21:]))
    else:
        raise ValueErro(f'Status {status!r} desconhecido')
df_results = pd.DataFrame(list_results).T
df_results.head()
```

```python
df_results2 = pd.concat([df_results.drop(['params'], axis=1), df_results['params'].apply(pd.Series)], axis=1).copy()
df_results2.head()
```

```python
df_results3 = pd.concat([df_results2['results'].apply(pd.Series), df_results2.drop(['results'], axis=1)], axis=1).copy()
df_results3.head()
```

```python
best_estimators = df_results3.sort_values(by='mean', ascending=False)
best_estimators.head()
```
